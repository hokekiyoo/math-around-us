## 大圏航路
北朝鮮からサンフランシスコにミサイルを打つ時、実は日本は通らない

球座標の緯度を、一階球状ベクトルで内分して、元に戻す
C = kR1 + (1-k)R2
- 緯度 theta
- 経度 phi



```python
def make_pattern(c):
    # 点がどの象限に存在するか
    if c[0]>0 and c[1]>0 and c[2]>=0:
        pattern = 0
    if c[0]>0 and c[1]>0 and c[2]<0:
        pattern = 1
    if c[0]>0 and c[1]<0 and c[2]>=0:
        pattern = 2
    if c[0]>0 and c[1]<0 and c[2]<0:
        pattern = 3
    if c[0]<0 and c[1]>0 and c[2]>=0:
        pattern = 4
    if c[0]<0 and c[1]>0 and c[2]<0:
        pattern = 5
    if c[0]<0 and c[1]<0 and c[2]>=0:
        pattern = 6
    if c[0]<0 and c[1]<0 and c[2]<0:
        pattern = 7
    return pattern

theta1, phi1 = [39,129] #平壌
theta2, phi2 = [33,-118] #Los Angels
# print(point1,point2)
theta1 = theta1*np.pi/180
phi1 = phi1*np.pi/180
theta2 = theta2*np.pi/180
phi2 = phi2*np.pi/180
ks = np.linspace(0,1,101)
C = np.zeros((len(ks),3))
# 起点、終点を中心を原点とするベクトルであらわす
R1 = r*np.array([np.cos(theta1)*np.cos(phi1),np.cos(theta1)*np.sin(phi1), np.sin(theta1)])
R2 = r*np.array([np.cos(theta2)*np.cos(phi2),np.cos(theta2)*np.sin(phi2), np.sin(theta2)])
for i,k in enumerate(ks):
    C[i] = (1-k)*R1+k*R2#弦
    C[i] *= r/np.linalg.norm(C[i])
    
# 極座標での緯度経度に直す
thetas,phis = [],[]
tmp = phi1*180/np.pi

for c in C:
    pattern = make_pattern(c)
    phi = np.arctan(c[1]/c[0])
    theta = np.arccos(np.sqrt(c[0]**2+c[1]**2)/(np.sqrt(c[0]**2+c[1]**2+c[2]**2)))
    theta = theta*180/np.pi
    phi = phi*180/np.pi
    if pattern == 1:
        theta = -theta
    if pattern == 3:
        theta = -theta
    if pattern == 4:
        phi += 180 
    if pattern == 5:
        theta = -theta 
        phi += 180
    if pattern == 6:
        phi -= 180
    if pattern == 7:
        theta = -theta
        phi -= 180
    phis.append(phi)
    thetas.append(mercator(theta))
# プロット
fig = plt.figure(figsize=(10,10))
# ax = plt.axes()
plotmap(coordinates,"taiken dist", grid=True)
plt.plot(phis,thetas,".",color="black",alpha=0.5)
plt.savefig("figures/5-6_great-circle-route.pdf")


```
