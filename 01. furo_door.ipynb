## お風呂のパカパカドア
ドアの開閉に係る通過領域を探る！

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("dark", {'grid.linestyle': '--'})
```

```python
# ドア左右の軌跡
l = 1
def left(t):
    x = np.sin(t)
    y = np.cos(t)
    return(x,y)
def right(t):
    x = 2*np.sin(t)
    y = 0
    return (x,y)
```

```python
# 通過領域の描画
ts = np.linspace(0,np.pi/2,100)
plt.figure(figsize=(10,5))
plt.xlim(0,2)
plt.ylim(0,1)
for t in ts:
    xl,yl = left(t)
    xr,yr = right(t)
    plt.plot([0,xl,xr],[0,yl,yr],"o-",color="blue",alpha=0.1)
plt.grid()
plt.savefig("figures/1-1_door.pdf")
```

## アステロイドと重ねる
- 青 : アステロイド
- 赤 : ドアのぱかぱか

```python
def asteroid_left(t):
    x = 0
    y = 2*np.cos(t)
    return(x,y)

ts = np.linspace(0,np.pi/2,200)
plt.figure(figsize=(10,10))
plt.xlim(0,2)
plt.ylim(0,2)
for t in ts:
    xl,yl = asteroid_left(t)
    xr,yr = right(t)
    plt.plot([0,xl,xr],[0,yl,yr],"-",color="blue",alpha=0.01)
    xl2,yl2 = left(t)
    plt.plot([0,xl2,xr],[0,yl2,yr],"-o",color="red",alpha=0.1)
plt.grid()
plt.savefig("figures/1-2_asteroid.pdf")
```
